### Objective
To apply statistical techniques on student performance data and use the results to support decision-making.

## Import Required Libraries

```python
import pandas as pd
```

## Load the Dataset

```python
df = pd.read_csv("student_data.csv")
df
```

## Statistical Summary

```python
df.describe()
```

## Calculate Important Statistics

```python
print("Mean Marks:", df["Marks"].mean())
print("Median Marks:", df["Marks"].median())
print("Maximum Marks:", df["Marks"].max())
print("Minimum Marks:", df["Marks"].min())
print("Standard Deviation:", round(df["Marks"].std(),2))
```

## Students Scoring Above Average

```python
average = df["Marks"].mean()

above_average = df[df["Marks"] > average]

above_average
```

## Decision Support

```python
average_marks = df["Marks"].mean()
attendance = df["Attendance"].mean()

print("Average Marks:", round(average_marks,2))
print("Average Attendance:", round(attendance,2))

if average_marks >= 85:
    print("\nDecision: Student performance is Excellent.")
elif average_marks >= 75:
    print("\nDecision: Student performance is Good, but can be improved.")
else:
    print("\nDecision: Additional academic support is required.")
```

## Conclusion

Statistical analysis helps convert raw data into meaningful insights. By calculating measures such as mean, median, and standard deviation, institutions can make informed decisions regarding student performance and academic improvement.
