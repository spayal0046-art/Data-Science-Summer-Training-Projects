## Business Problem

A telecom company is losing customers every month because many customers switch to other service providers.

This results in:
- Loss of revenue
- Increased marketing cost
- Reduced customer satisfaction

## Business Objective

The objective is to predict which customers are likely to leave the company so that special offers and retention strategies can be provided to them.

## Required Dataset

The following data is required for this project:

| Column Name | Description |
|-------------|-------------|
| Customer_ID | Unique customer ID |
| Age | Customer age |
| Gender | Male/Female |
| Monthly_Charges | Monthly bill amount |
| Contract_Type | Monthly/Yearly |
| Tenure | Number of months with the company |
| Churn | Yes/No |

## Machine Learning Problem

**Problem Type:** Classification

The model will predict whether a customer is likely to leave the company (Churn = Yes) or stay (Churn = No).

## Expected Outcome

- Identify customers likely to leave
- Improve customer retention
- Increase company revenue
- Support better business decisions
