### Objective
Prepare raw data for analysis by identifying and handling missing values, duplicate records, and inconsistencies to create a clean dataset.

## Importing Required Libraries

```python
import pandas as pd
```

## Loading the Dataset

```python
df = pd.read_csv("student_data.csv")
df
```

## Inspect Dataset Structure

```python
df.info()
```

## Checking Missing Values

```python
df.isnull().sum()
```

## Removing Duplicate Records

```python
df = df.drop_duplicates()
df
```

## Handling Missing Values

```python
df["Name"] = df["Name"].fillna("Unknown")

df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
```

## Clean Dataset

```python
df
```

## Verifying Clean Data

```python
df.isnull().sum()
```

## Conclusion

The raw dataset was successfully cleaned by:

- Removing duplicate records
- Handling missing values
- Inspecting dataset structure
- Preparing a clean, analysis-ready dataset

This cleaned dataset can now be used for data analysis and Machine Learning.
